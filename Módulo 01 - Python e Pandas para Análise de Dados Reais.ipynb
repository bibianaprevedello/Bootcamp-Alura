```python
import numpy as np
np.random.seed(687942)

usaveis.sample(n=7)
```

```python
usaveis["Total"] = usaveis.sum(axis=1)
usaveis.head()
```

```python
ordenado_por_total = usaveis.sort_values("Total", ascending=False)
ordenado_por_total = ordenado_por_total.drop("Total", axis=1)
ordenado_por_total.head()
```

```python
ordenado_por_total.columns
```

```python
colunas_interessadas = ordenado_por_total.columns[6:]
ordenado_por_total = ordenado_por_total[colunas_interessadas]
```

```python
ordenado_por_total.head(5).T.plot(figsize=(10,6))
```

### Desafios:

- Plotar só os estados do sudeste e ver se afetou com um pico em 2012/2013 de forma similar esses estados

```python
ordenado_por_total.head()
```

```python
regiao_sudeste = ordenado_por_total

regiao_sudeste["Região"] = ordenado_por_total.index.str[0]

dicionario_regiao = {"1": "Norte", "2": "Nordeste", "3": "Sudeste", "4": "Sul", "5": "Centro Oeste"}

for item in dicionario_regiao.keys():
    regiao_sudeste["Região"] = regiao_sudeste["Região"].replace(item, dicionario_regiao[item])
    
regiao_sudeste.head()
```

```python
regiao_sudeste = regiao_sudeste[regiao_sudeste.Região == "Sudeste"]
regiao_sudeste
```

```python
regiao_sudeste = regiao_sudeste.drop("Região", axis=1)
regiao_sudeste
```

```python
regiao_sudeste.T.plot(figsize=(10,6))
```

- Adicione seu estado aos 5 estados existentes e plote (Santa Catarina)

```python
santa_catarina = ordenado_por_total.T[["42 Santa Catarina"]]
santa_catarina

top_5_mais_santa_catarina = ordenado_por_total.head(5).T
top_5_mais_santa_catarina

top_5_mais_santa_catarina["42 Santa Catarina"] = santa_catarina
top_5_mais_santa_catarina
```

```python
ax4 = top_5_mais_santa_catarina.plot(figsize=(10,6))
ax4.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.title("Gasto dos top 5 estados + Santa Catarina")
ax4.legend(["São Paulo", "Minas Gerais", "Paraná", "Rio Grande do Sul", "Rio de Janeiro", "Santa Catarina"])
```
