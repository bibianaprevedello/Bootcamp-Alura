## 01. Aquisição e Leitura de Dados Reais

```python
import pandas as pd

dados = pd.read_csv("A054506189_28_143_208.csv", encoding="ISO-8859-1",
                   skiprows=3, sep=";", skipfooter=12, thousands=".", decimal=",",  engine="python")
dados.head()
```

```python
dados.tail()
```

```python
dados.info()
```

```python
pd.options.display.float_format = "{:.2f}".format
```

```python
dados.mean()
```

```python
dados["2008/Ago"]
```

```python
dados.head()
```

```python
dados["2008/Ago"].mean()
```

```python
dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(9,6))
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

ax = dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(9,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
                             
plt.title("Valor por unidade da Federação")
```

### Desafios:

- Escolher um título mais descritivo, que passe a mensagem adequada
